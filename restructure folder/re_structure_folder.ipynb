```python
import os
import pathlib
from pathlib import Path, PureWindowsPath
import shutil

```

```python
def countslash(str1):
    count = 0 
    for i in range(len(str1)):
        if str1[i]== '\\':
            count = count + 1
    return count
```

```python
def getsubfolder(f):
    for fder in os.scandir(f):
        sf = os.path.abspath(fder)
        if countslash(sf) == 9:
            p = Path(sf).parent.absolute()
            p1 =  os.path.abspath(p)
            # print(p)
            sff = os.path.basename(sf)
            # print(sff)
            print(f'fullpath la : {p}')
            print(f'filename la : {sff}')
            client = os.path.basename(p)
            month = os.path.basename(p.parents[0])
            year = os.path.basename(p.parents[1])
            ins = os.path.basename(p.parents[2])
            newdir = 'C:/Users/u1112180/Downloads/test_client1/' + client + '/' + year + '/' + month + '/' + ins 
            print(newdir) 
            if os.path.exists(newdir):
                print("folder tontai")
                shutil.copyfile(p1+ "/" + sff, newdir + '/' + sff)
            else:
                os.makedirs(newdir)
                print("chua co folder ")
                shutil.copyfile(p1 + "/" + sff, newdir + '/' + sff)                      
        if os.path.isdir(os.path.abspath(sf)):
            getsubfolder(os.path.abspath(sf))
# p11 = r'C:\Users\u1112180\Downloads\test_structure1'
p11 = 'C:/Users/u1112180/Downloads/test_structure1'
getsubfolder(p11)
```
