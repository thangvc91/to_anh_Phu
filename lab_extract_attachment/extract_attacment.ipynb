```python
def saveattachemnts(message):
    #TODO: 
    # get subject
    # list subject -> ghep path output 
    # check if folder exist -> save as attachment
    sb = message.Subject
    sb = sb[sb.find("|")+1:].replace("_", "-").replace("Hieu luc", "")
    l = sb.split("-")
    l = [x.strip() for x in l ] #['BVIET', 'BV Dong Do', 'Akzo', 'BCBT', '08 2020', 'Full year']
    ins1 = l[0] #'BVIET'
    y = l[4][-4:] #'2020'
    m = l[4][:2] #'08'
    client1 = l[2] #'Akzo'
    p = basepath + "/" + ins1 + "/" + y + "/" + m + "/" + client1
    attachments = message.Attachments
    attachment = attachments.Item(1)
    f   = 0
    for attachment in message.Attachments:
        if os.path.exists(p):
            attachment.SaveAsFile(p + "/" + str(attachment))
        else:
            os.makedirs(p)
            attachment.SaveAsFile(p + "/" + str(attachment))
                
for mes in messages:
    saveattachemnts(mes)
```
