```python
#define function 
def listconvert(df_test):
    df1 = []
    for i in df_test:
        if str(i) != 'nan':
            df1.append(i)
    return df1
```

```python
today = datetime.now().strftime("%d-%m-%Y")
def backup(path):
    f=0
    if os.path.isdir(path):
        if len(os.listdir(path)) > 0:
            if os.path.exists(path + "/000_BACKUP"):
                f=1
            else:
                os.makedirs(path + "/000_BACKUP")
                f = 1
            sf = [g for g in os.listdir(path)]
            if f ==1:
                for sff in sf:
                    if not os.path.isdir(path + "/" + sff):
                        shutil.move(path + "/" + sff, path + "/000_BACKUP/" + today + "__" + sff )            
             
```
