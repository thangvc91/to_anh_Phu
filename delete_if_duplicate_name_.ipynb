```python
def modification_date(filename):
    t = os.path.getmtime(filename)
    return datetime.datetime.fromtimestamp(t)
```

```python
def deleteoldfile(file1,file2):
    t1 = modification_date(file1)
    t2 = modification_date(file2)
    if t1 > t2: 
        os.remove(file2)
    else:
        os.remove(file1)
```
